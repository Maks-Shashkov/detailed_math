Провести статистическую обработку текста.
Определить энтропию, приходящуюся в среднем на одну букву, длину кода при
равномерном кодировании и избыточность.

Функция calculate_entropy вычисляет энтропию для заданного текста. Она использует класс Counter из модуля collections, чтобы подсчитать частоту каждого символа в тексте. Затем она вычисляет вероятность каждого символа и применяет формулу Шеннона для вычисления энтропии. Результатом функции является значение энтропии.

Функция calculate_average_code_length вычисляет среднюю длину кода при равномерном кодировании для заданного текста. Она также использует класс Counter для подсчета частоты каждого символа в тексте. Затем она вычисляет вероятность каждого символа и применяет формулу Хартли для вычисления средней длины кода. Результатом функции является средняя длина кода.

Функция calculate_redundancy вычисляет избыточность для заданного текста. Она вызывает функции calculate_entropy и calculate_average_code_length, чтобы вычислить энтропию и среднюю длину кода. Затем она вычитает энтропию из средней длины кода, чтобы получить избыточность. Результатом функции является значение избыточности.

В примере использования кода, переменная text содержит текст, который мы хотим обработать. Мы вызываем каждую из функций (calculate_entropy, calculate_average_code_length, calculate_redundancy) для этого текста и сохраняем результаты в соответствующих переменных (entropy, average_code_length, redundancy).

Затем мы выводим значения энтропии, средней длины кода при равномерном кодировании и избыточности на экран с помощью функции print.

```python
import collections
import math

def calculate_entropy(text):
    freq = collections.Counter(text)
    total = len(text)
    entropy = 0.0
    for count in freq.values():
        probability = count / total
        entropy += probability * math.log2(probability)
    entropy = -entropy
    return entropy

def calculate_average_code_length(text):
    freq = collections.Counter(text)
    total = len(text)
    code_length = 0.0
    for count in freq.values():
        probability = count / total
        code_length += probability * math.log2(1/probability)
    return code_length

def calculate_redundancy(text):
    entropy = calculate_entropy(text)
    average_code_length = calculate_average_code_length(text)
    redundancy = average_code_length - entropy
    return redundancy

# Пример использования:
text = "Текст для примера"
entropy = calculate_entropy('ьлавыпаптощвтпоавпатов')
average_code_length = calculate_average_code_length('ьлавыпаптощвтпоавпатов')
redundancy = calculate_redundancy('ьлавыпаптощвтпоавпатов')

print("Энтропия:", entropy)
print("Средняя длина кода при равномерном кодировании:", average_code_length)
print("Избыточность:", redundancy)
```

Далее определяется функция build_huffman_tree, которая принимает текст и возвращает корень дерева Хаффмана для символов в тексте. Затем создается очередь с приоритетом из узлов, каждый из которых содержит символ и его частоту. Очередь сортируется по возрастанию частоты.

Функция рекурсивно обходит дерево, начиная с корня, и назначает код '0' для левого поддерева и код '1' для правого поддерева. Когда функция достигает листового узла, она добавляет символ и его код в таблицу кодов.

Определяются также функции encode_text и decode_text, которые соответственно кодируют текст с использованием таблицы кодов Хаффмана и декодируют закодированный текст с использованием дерева Хаффмана.

В конце кода определен пример использования всех функций для заданного текста. Функция build_huffman_tree используется для построения дерева Хаффмана для данного текста. Затем функция build_huffman_code_table используется для построения таблицы кодов Хаффмана для каждого символа в дереве. Функция encode_text используется для закодирования текста с использованием таблицы кодов Хаффмана, а функция decode_text используется для декодирования закодированного текста с использованием дерева Хаффмана.

В конце кода вычисляется средняя длина элементарного кода и эффективность сжатия.

```python
import heapq
from collections import Counter

class Node:
    def __init__(self, char, freq):
        self.char = char
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(text):
    # Подсчет частоты символов в тексте
    freq_table = Counter(text)
    # Создание очереди с приоритетом из узлов
    priority_queue = [Node(char, freq) for char, freq in freq_table.items()]
    heapq.heapify(priority_queue)

    while len(priority_queue) > 1:
        # Извлечение двух узлов с наименьшей частотой
        node1 = heapq.heappop(priority_queue)
        node2 = heapq.heappop(priority_queue)
        # Создание нового узла с объединенной частотой
        merged_node = Node(None, node1.freq + node2.freq)
        merged_node.left = node1
        merged_node.right = node2
        # Добавление нового узла в очередь
        heapq.heappush(priority_queue, merged_node)

    # Возвращение корня дерева Хаффмана
    return priority_queue[0]

def build_huffman_code_table(node, code='', code_table={}):
    if node is None:
        return

    if node.char is not None:
        code_table[node.char] = code

    build_huffman_code_table(node.left, code + '0', code_table)
    build_huffman_code_table(node.right, code + '1', code_table)

    return code_table

def encode_text(text, code_table):
    encoded_text = ''
    for char in text:
        encoded_text += code_table[char]
    return encoded_text

def decode_text(encoded_text, huffman_tree):
    decoded_text = ''
    current_node = huffman_tree
    for bit in encoded_text:
        if bit == '0':
            current_node = current_node.left
        else:
            current_node = current_node.right

        if current_node.char is not None:
            decoded_text += current_node.char
            current_node = huffman_tree

    return decoded_text

# Пример использования кода

text = "Таким образом, новая модель организационной деятельности обеспечивает широкому кругу специалистов участие в формировании модели развития. Задача организации, в особенности же постоянный количественный рост и сфера нашей активности в значительной степени обуславливает создание форм воздействия! Равным образом повышение уровня гражданского сознания влечет за собой процесс внедрения и модернизации позиций, занимаемых участниками в отношении поставленных задач."

huffman_tree = build_huffman_tree(text)
code_table = build_huffman_code_table(huffman_tree)

encoded_text = encode_text(text, code_table)
decoded_text = decode_text(encoded_text, huffman_tree)

print("Закодированный текст:", encoded_text)
print("Декодированный текст:", decoded_text)

# Вычисление средней длины элементарного кода и эффективности сжатия
avg_code_length = sum(len(code) for code in code_table.values()) / len(code_table)
compression_ratio = len(encoded_text) / (len(text) * 8)  # Предполагаем, что каждый символ кодируется 8 битами

print("Средняя длина элементарного кода:", avg_code_length)
print("Эффективность сжатия:", compression_ratio)
```

Первые несколько строк кода определяют класс Node, который представляет узел дерева Хаффмана. Каждый узел содержит двухбуквенное сочетание (chars) и его частоту (freq). Класс также определяет метод lt для сравнения узлов по их частоте.

Далее определяется функция build_huffman_tree, которая принимает текст и возвращает корень дерева Хаффмана для двухбуквенных сочетаний в тексте. Функция начинается с создания таблицы частот для всех двухбуквенных сочетаний в тексте с помощью Counter из модуля collections. Затем создается очередь с приоритетом из узлов, каждый из которых содержит двухбуквенное сочетание и его частоту. Очередь сортируется по возрастанию частоты.

Далее функция выполняет следующие действия, пока в очереди не останется только один узел:

Извлекает два узла с наименьшей частотой из очереди.
Создает новый узел, объединяя два извлеченных узла, и присваивает ему частоту, равную сумме частот извлеченных узлов.
Добавляет новый узел в очередь.
В конце функция возвращает корень дерева Хаффмана.

Далее определяется функция build_huffman_code_table, которая принимает корень дерева Хаффмана и возвращает таблицу кодов Хаффмана для каждого двухбуквенного сочетания в дереве. Функция рекурсивно обходит дерево, начиная с корня, и строит коды для каждого двухбуквенного сочетания на основе пути от корня до листового узла, где сочетание находится.

Наконец, определяются функции encode и decode, которые принимают текст и таблицу кодов Хаффмана и возвращают закодированный или декодированный текст соответственно.

В целом, этот код реализует алгоритм Хаффмана для двухбуквенных сочетаний в тексте и позволяет закодировать и декодировать текст на основе построенной схемы кодирования.

```python
import heapq
from collections import Counter

class Node:
    def __init__(self, chars, freq):
        self.chars = chars
        self.freq = freq
        self.left = None
        self.right = None

    def __lt__(self, other):
        return self.freq < other.freq

def build_huffman_tree(text):
    # Подсчет частоты двухбуквенных сочетаний в тексте
    freq_table = Counter(text[i:i+2] for i in range(len(text)-1))
    # Создание очереди с приоритетом из узлов
    priority_queue = [Node(chars, freq) for chars, freq in freq_table.items()]
    heapq.heapify(priority_queue)

    while len(priority_queue) > 1:
        # Извлечение двух узлов с наименьшей частотой
        node1 = heapq.heappop(priority_queue)
        node2 = heapq.heappop(priority_queue)
        # Создание нового узла с объединенной частотой
        merged_node = Node(None, node1.freq + node2.freq)
        merged_node.left = node1
        merged_node.right = node2
        # Добавление нового узла в очередь
        heapq.heappush(priority_queue, merged_node)

    # Возвращение корня дерева Хаффмана
    return priority_queue[0]

def build_huffman_code_table(node, code='', code_table={}):
    if node is None:
        return

    if node.chars is not None:
        code_table[node.chars] = code

    build_huffman_code_table(node.left, code + '0', code_table)
    build_huffman_code_table(node.right, code + '1', code_table)

    return code_table

def encode_text(text, code_table):
    encoded_text = ''
    i = 0
    while i < len(text)-1:
        chars = text[i:i+2]
        if chars in code_table:
            encoded_text += code_table[chars]
            i += 2
        else:
            encoded_text += text[i]
            i += 1

    if i == len(text)-1:
        encoded_text += text[i]

    return encoded_text

def decode_text(encoded_text, code_table):
    decoded_text = ''
    current_code = ''
    for bit in encoded_text:
        current_code += bit
        if current_code in code_table:
            decoded_text += code_table[current_code]
            current_code = ''

    return decoded_text

text = "Таким образом, новая модель организационной деятельности обеспечивает широкому кругу специалистов участие в формировании модели развития. Задача организации, в особенности же постоянный количественный рост и сфера нашей активности в значительной степени обуславливает создание форм воздействия! Равным образом повышение уровня гражданского сознания влечет за собой процесс внедрения и модернизации позиций, занимаемых участниками в отношении поставленных задач."

# Построение схемы алфавитного кодирования для двухбуквенных сочетаний
huffman_tree = build_huffman_tree(text)
code_table = build_huffman_code_table(huffman_tree)

# Закодирование текста
encoded_text = encode_text(text, code_table)

# Декодирование текста
decoded_text = decode_text(encoded_text, code_table)

print("Схема алфавитного кодирования:")
for chars, code in code_table.items():
    print(f"{chars}: {code}")

print("\nЗакодированный текст:")
print(encoded_text)

print("\nДекодированный текст:")
print(decoded_text)

# Вычисление средней длины элементарного кода
average_code_length = sum(len(code) for code in code_table.values()) / len(code_table)

# Вычисление эффективности сжатия
original_length = len(text) * 8  # Предполагаем, что каждый символ кодируется 8 битами
compressed_length = len(encoded_text)
compression_ratio = original_length / compressed_length

print(f"\nСредняя длина элементарного кода: {average_code_length} бит")
print(f"Эффективность сжатия: {compression_ratio:.2f}")
```
